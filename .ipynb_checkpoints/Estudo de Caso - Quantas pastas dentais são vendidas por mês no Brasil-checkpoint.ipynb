# Calculo de consumo:

### Calculo da MASSA (g) pela DENSIDADE (g/mm³) e VOLUME (mm³)

```python
densidade_pasta = 1.6/1000
densidade_pasta
```

#### mm²

```python
area_do_bico_saida = (3.14 * 5**2)
area_do_bico_saida
```

#### mm³

```python
volume_de_pasta_na_escova = media_tamanho_cabeca_escova * area_do_bico_saida
volume_de_pasta_na_escova
```

#### m = v * d 

```python
massa_pasta_na_escova = volume_de_pasta_na_escova * densidade_pasta
massa_pasta_na_escova
```

#### Consumo de pasta pela população com acesso a higiene bucal

```python
consumo_pasta_dental_ufmg
```

#### Retirando as pessoas que possuem banheiro mas não utilizam pasta de dente

```python
pessoas_que_consomem_pasta_dentaria = consumo_pasta_dental_ufmg * estimativa_atual_de_pessoas_com_banheiro
pessoas_que_consomem_pasta_dentaria
```

#### Média de 1,5x escovações por dia

```python
frequencia_diaria_escovacao_1a2x_ufrs 
```

```python
numero_diario_escovacao_por_pessoa_1a2x_ufrs = (pessoas_que_consomem_pasta_dentaria * frequencia_diaria_escovacao_1a2x_ufrs)*1.5
numero_diario_escovacao_por_pessoa_1a2x_ufrs
```

#### Média de 2,5x escovações por dia

```python
frequencia_diaria_escovacao_2a3x_ufrs
```

```python
numero_diario_escovacao_por_pessoa_2a3x_ufrs = (pessoas_que_consomem_pasta_dentaria * frequencia_diaria_escovacao_2a3x_ufrs)*2.5
arredondar(numero_diario_escovacao_por_pessoa_2a3x_ufrs)
```

#### Total de escovações por pessoa no dia

```python
total_diario_escovacao_por_pessoa = numero_diario_escovacao_por_pessoa_1a2x_ufrs + numero_diario_escovacao_por_pessoa_2a3x_ufrs
arredondar(total_diario_escovacao_por_pessoa)
```

# Resultado Final

```python
total_tubos_mes = ((massa_pasta_na_escova * total_diario_escovacao_por_pessoa)/90) * 30
arredondar(total_tubos_mes)
```

```python
print(f"São vendidas por mês no Brasil {arredondar(total_tubos_mes)} pastas dentais") 
```

## Tubos por pessoa mês:

```python
tubos_pessoa_mes = total_tubos_mes / pessoas_que_consomem_pasta_dentaria
arredondar(tubos_pessoa_mes)
```
